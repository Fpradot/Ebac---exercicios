````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
````

````python
df = pd.read_csv('Profissao- Cientista de Dados_M10_support material.csv')
df.head()
````

````python
df1 = df[['data_ref','sexo','posse_de_veiculo','posse_de_imovel','tipo_renda','educacao','estado_civil','tipo_residencia']]
````

````python
df2 = df1.copy()
df2.loc[:, ('data_ref')] = pd.to_datetime(df2['data_ref'])
````

1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada?

````python
ax = sns.countplot(x='data_ref', hue='sexo', data=df2)
tick_labs = df2['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()
ticks = ax.set_xticks(list(range(df2['data_ref'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation = 90)
plt.legend(bbox_to_anchor=(1.0,1),loc=2,borderaxespad=0.)
````

````python
ax = sns.countplot(x='data_ref', hue='posse_de_imovel', data=df2)
tick_labs = df2['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()
ticks = ax.set_xticks(list(range(df2['data_ref'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation = 90)
plt.legend(bbox_to_anchor=(1.05,1),loc=2,borderaxespad=0.)
````

````python
ax = sns.countplot(x='data_ref', hue='tipo_renda', data=df2)
tick_labs = df2['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()
ticks = ax.set_xticks(list(range(df2['data_ref'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation = 90)
plt.legend(bbox_to_anchor=(1.05,1),loc=2,borderaxespad=0.)
````

````python
ax = sns.countplot(x='data_ref', hue='educacao', data=df2)
tick_labs = df2['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()
ticks = ax.set_xticks(list(range(df2['data_ref'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation = 90)
plt.legend(bbox_to_anchor=(1.05,1),loc=2,borderaxespad=0.)
````

- Houve um aumento significativo de pessoas com superior completo a partir do 09-2015.

````python
ax = sns.countplot(x='data_ref', hue='estado_civil', data=df2)
tick_labs = df2['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()
ticks = ax.set_xticks(list(range(df2['data_ref'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation = 90)
plt.legend(bbox_to_anchor=(1.05,1),loc=2,borderaxespad=0.)
````

- No mês 11-2015 houve um aumento de união.

2) Para cada variável qualitativa, construa um gráfico mostrando as médias de cada nível da variável no tempo e avalie a estabilidade. Houve alguma alteração importante? Você acha que se pode utilizar essas variáveis em uma modelagem?

````python
tab = pd.crosstab(df2['data_ref'],df2['sexo'])
tabMed = tab.div(tab.sum(axis=1), axis =0)
ax = tabMed.plot.bar(stacked=True)
tick_labs = df2['data_ref'].map(lambda ts: ts.strftime("%m-%Y")).unique()
ticks = ax.set_xticks(list(range(df2['data_ref'].nunique())))
labels = ax.set_xticklabels(tick_labs, rotation = 90)
plt.legend(bbox_to_anchor=(1.05,1),loc=2,borderaxespad=0.)
````
